Platform_c nin Campaign name ilişkisine göre koşullu mod ile doldurulması

```python
# Adım 1: Kırılma noktasını düzeltiyoruz – to_frame'e özel isim veriyoruz
campaign_platform_ratio = (
    df[df['platform__c'].notnull()]
    .groupby(['campaign.name', 'platform__c'])
    .size()
    .reset_index(name='count')  # Bu artık tekrar etmeyen isim olacak
)

# Adım 2: Toplam kampanya sayısına göre oranları hesapla
total_counts = campaign_platform_ratio.groupby('campaign.name')['count'].transform('sum')
campaign_platform_ratio['ratio'] = campaign_platform_ratio['count'] / total_counts
```

```python
campaign_platform_ratio
```

```python
# %90 üzeri eşleşmeleri al
strong_matches = campaign_platform_ratio[campaign_platform_ratio['ratio'] >= 0.9]

# Sözlük oluştur: {'Campaign.Name': 'Platform__c'}
campaign_to_platform = dict(zip(strong_matches['campaign.name'], strong_matches['platform__c']))
```

```python
campaign_to_platform
```

```python
from tqdm import tqdm

for i in tqdm(df[df['platform__c'].isnull()].index):

    campaign = df.loc[i, 'campaign.name']
    recordtype = df.loc[i, 'recordtype.name']

    # 1. Kampanyaya göre doldur
    if campaign in campaign_to_platform:
        df.at[i, 'platform__c'] = campaign_to_platform[campaign]

    # 2. Kampanya eşleşmedi, record type'a göre mod
    elif pd.notnull(recordtype):
        platform_mode = df[df['recordtype.name'] == recordtype]['platform__c'].mode()
        if not platform_mode.empty:
            df.at[i, 'platform__c'] = platform_mode[0]

    # 3. Hiçbir şey yoksa genel mod
    else:
        overall_mode = df['platform__c'].mode()
        if not overall_mode.empty:
            df.at[i, 'platform__c'] = overall_mode[0]
```

```python
def calculate_LB_UB (df,var):
    '''calculating upper bound and lower bound values to identify outliers'''
    Q3=np.percentile(df[var],75)
    Q1=np.percentile(df[var],25)
    IQR=1.5*(Q3-Q1)
    
    UB=Q3+IQR
    LB=Q1-IQR
    
    print(LB,UB)
    
    return LB,UB
```

```python
#numerical_columns = df.select_dtypes('float64', 'int').columns

for i in num_cols:
    print("\n name of column:", i)
    LB, UB = calculate_LB_UB(df, i)
    print("min and max value", (df[i].min(), df[i].max()))
    print("range of outliers:", LB, UB)
    
    if df[i].max() > UB or df[i].min() < LB:
        print("----Investigate variable for outliers---")
```

```python
# Hedef Değişken Yaratımı
df['donated'] = np.where(df['recordtype.name'] == 'Donation', 1, 0)
```

```python
# Tekrar Sipariş (Repeat Customer)
repeat_customers = df.groupby('contactid').size() > 1
df['repeatcustomer'] = df['contactid'].map(repeat_customers).fillna(False).astype(int)
```

```python

# Platformları One-Hot Encoding
df = pd.get_dummies(df, columns=['platform__c', 'campaign.name'], drop_first=True)
```

```python
# 3. Feature Engineering
total_amount = df.groupby('contactid')['amount'].sum().rename('totalamount')
total_transactions = df.groupby('contactid').size().rename('totaltransactions')
first_transaction = df.groupby('contactid')['closedate'].min().rename('firsttransactiondate')
last_transaction = df.groupby('contactid')['closedate'].max().rename('lasttransactiondate')
today = pd.Timestamp('today')
days_since_last_transaction = (today - last_transaction).dt.days.rename('dayssincelasttransaction')
campaign_participation = df.groupby('contactid')['campaign.name_No campaign'].nunique().rename('campaignparticipationcount')
donation_counts = df[df['donated'] == 1].groupby('contactid').size()
donation_rate = (donation_counts / total_transactions).fillna(0).rename('donationrate')
```

```python
# Kullanıcı bazlı özellikleri birleştirme
features_df = pd.concat([
    total_amount,
    total_transactions,
    first_transaction,
    last_transaction,
    days_since_last_transaction,
    campaign_participation,
    donation_rate
], axis=1)

features_df.reset_index(inplace=True)
```

```python
features_df.head()
```

```python

# 4. Modelleme için Veri Hazırlama
model_data = features_df.copy()
model_data.fillna(0, inplace=True)  # Güvenlik için boşları doldur

# Bağış yapma durumunu ekleyelim
contact_donated = df[['contactid', 'Donated']].drop_duplicates(subset='contactid')
model_data = model_data.merge(contact_donated, how='left', on='contactid')
model_data['Donated'].fillna(0, inplace=True)

# X ve y oluştur
X = model_data.drop(['contactid', 'firsttransactiondate', 'lasttransactiondate', 'Donated'], axis=1)
y = model_data['Donated']

# Standard Scaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# 5. Logistic Regression
lr = LogisticRegression(max_iter=2000)
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)

print("Logistic Regression Sonuçları:")
print(classification_report(y_test, y_pred_lr))
print("ROC AUC:", roc_auc_score(y_test, lr.predict_proba(X_test)[:,1]))
```

```python
# 1. Random Forest Modeli
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)

# 2. Performans Raporu
print("\nRandom Forest Sonuçları:")
print(classification_report(y_test, y_pred_rf))
print("ROC AUC:", roc_auc_score(y_test, rf.predict_proba(X_test)[:,1]))

# 3. Confusion Matrix Çizimi
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, y_pred_rf)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap='Blues')
plt.title('Random Forest Confusion Matrix')
plt.show()
```

```python
# 7. Propensity Skorları
model_data['PropensityScore_RF'] = rf.predict_proba(X_scaled)[:,1]

# 8. Churn Risk (DaysSinceLastTransaction > 365 günü aşmış kullanıcılar)
model_data['ChurnRisk'] = np.where(model_data['dayssincelasttransaction'] > 365, 1, 0)

```

```python

# 2. RFM Analizi
# RFM Analizi düzeltilmiş kod
def rfm_analysis(df):
    # Analysis date will be one day after the max date
    analysis_date = df['closedate'].max() + pd.Timedelta(days=1)
    
    # RFM metriklerini hesapla
    rfm = df.groupby('contactid').agg({
        'closedate': lambda x: (analysis_date - x.max()).days,  # Recency
        'opp. id': 'count',  # Frequency
        'amount': 'sum'  # Monetary
    }).reset_index()
    
    rfm.columns = ['ContactId', 'Recency', 'Frequency', 'Monetary']
    
    # RFM Scoring
    r_labels = range(4, 0, -1)
    f_labels = range(1, 5)
    m_labels = range(1, 5)
    
    # Quartile hesapları ve etiketleme
    try:
        r_quartiles = pd.qcut(rfm['Recency'], q=4, labels=r_labels, duplicates='drop')
        f_quartiles = pd.qcut(rfm['Frequency'], q=4, labels=f_labels, duplicates='drop')
        m_quartiles = pd.qcut(rfm['Monetary'], q=4, labels=m_labels, duplicates='drop')
    except ValueError:
        # Eğer quartile hesabı başarısız olursa manuel bölümleme yap
        r_quartiles = pd.cut(rfm['Recency'], bins=4, labels=r_labels)
        f_quartiles = pd.cut(rfm['Frequency'], bins=4, labels=f_labels)
        m_quartiles = pd.cut(rfm['Monetary'], bins=4, labels=m_labels)
    
    rfm['R'] = r_quartiles
    rfm['F'] = f_quartiles
    rfm['M'] = m_quartiles
    
    # Eksik değerleri doldur
    rfm[['R', 'F', 'M']] = rfm[['R', 'F', 'M']].fillna(1)
    
    # RFM Score hesapla
    rfm['RFM_Score'] = rfm['R'].astype(str) + rfm['F'].astype(str) + rfm['M'].astype(str)
    
    # Müşteri segmentasyonu
    def segment_customers(row):
        if row['R'] >= 3 and row['F'] >= 3 and row['M'] >= 3:
            return 'Best Customers'
        elif row['R'] == 1:
            return 'Lost Customers'
        elif row['F'] >= 3 and row['M'] >= 3:
            return 'Loyal Customers'
        elif row['M'] >= 3:
            return 'Big Spenders'
        elif row['F'] >= 3:
            return 'Frequent Customers'
        else:
            return 'Average Customers'
    
    rfm['Customer_Segment'] = rfm.apply(segment_customers, axis=1)
    
    return rfm

# RFM analizini çalıştır
rfm_results = rfm_analysis(df)

# Sonuçları görselleştir
plt.figure(figsize=(10, 6))
segment_counts = rfm_results['Customer_Segment'].value_counts()
sns.barplot(x=segment_counts.index, y=segment_counts.values)
plt.title('Customer Segments Distribution')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Segment bazlı istatistikler
segment_stats = rfm_results.groupby('Customer_Segment').agg({
    'Recency': 'mean',
    'Frequency': 'mean',
    'Monetary': 'mean'
}).round(2)

print("\nSegment Bazlı Ortalama Değerler:")
print(segment_stats)

# 3. Cohort Analizi
def cohort_analysis(df):
    # Tarihleri datetime'a çevir
    df['closedate'] = pd.to_datetime(df['closedate'])
    
    # NaN değerleri olan satırları filtrele
    df = df.dropna(subset=['closedate', 'contactid'])
    
    # İlk işlem tarihini bul
    df['CohortMonth'] = df.groupby('contactid')['closedate'].transform('min').dt.to_period('M')
    
    # İşlem ayını bul
    df['TransactionMonth'] = df['closedate'].dt.to_period('M')
    
    # Cohort indeksini hesapla (ay farkı)
    df['CohortIndex'] = ((df['TransactionMonth'].astype(str).apply(lambda x: pd.Period(x)) - 
                         df['CohortMonth'].astype(str).apply(lambda x: pd.Period(x))).apply(lambda x: x.n))
    
    # Cohort tablosunu oluştur
    cohort_data = df.groupby(['CohortMonth', 'CohortIndex'])['contactid'].nunique().reset_index()
    cohort_table = cohort_data.pivot(index='CohortMonth', columns='CohortIndex', values='contactid')
    
    # Retention rates hesapla
    cohort_size = cohort_table[0]
    retention_table = cohort_table.divide(cohort_size, axis=0) * 100
    
    return retention_table

# 4. LTV (Lifetime Value) Analizi
def calculate_ltv(df):
    # Customer lifetime calculations
    customer_life = df.groupby('contactid').agg({
        'closedate': lambda x: (x.max() - x.min()).days / 365.25,  # Years
        'amount': 'sum',
        'opp. id': 'count'
    }).reset_index()
    
    customer_life.columns = ['ContactId', 'CustomerLife', 'TotalRevenue', 'TransactionCount']
    
    # Calculate annual metrics
    customer_life['AnnualRevenue'] = customer_life['TotalRevenue'] / customer_life['CustomerLife']
    customer_life['TransactionsPerYear'] = customer_life['TransactionCount'] / customer_life['CustomerLife']
    customer_life['AverageTransactionValue'] = customer_life['TotalRevenue'] / customer_life['TransactionCount']
    
    # Calculate LTV
    customer_life['LTV'] = customer_life['AnnualRevenue'] * customer_life['CustomerLife']
    
    return customer_life

# Analizleri çalıştır
rfm_results = rfm_analysis(df)
cohort_results = cohort_analysis(df)
ltv_results = calculate_ltv(df)

# Görselleştirmeler

# 1. RFM Segment Dağılımı
plt.figure(figsize=(10, 6))
segment_counts = rfm_results['Customer_Segment'].value_counts()
sns.barplot(x=segment_counts.index, y=segment_counts.values)
plt.title('Customer Segments Distribution')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 2. Cohort Retention Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(cohort_results, annot=True, fmt='.0f', cmap='YlOrRd')
plt.title('Customer Retention Rate by Cohort (%)')
plt.xlabel('Cohort Index (Months)')
plt.ylabel('Cohort Month')
plt.tight_layout()
plt.show()

# 3. LTV Dağılımı
plt.figure(figsize=(10, 6))
sns.histplot(data=ltv_results, x='LTV', bins=50)
plt.title('Customer Lifetime Value Distribution')
plt.xlabel('Lifetime Value ($)')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

# Özet İstatistikler
print("\nRFM Segment Summary:")
print(rfm_results['Customer_Segment'].value_counts())

print("\nAverage LTV Metrics:")
print(ltv_results[['AnnualRevenue', 'TransactionsPerYear', 'AverageTransactionValue', 'LTV']].describe())

print("\nCohort Retention Özeti:")
print("\nOrtalama Retention Oranları:")
print(cohort_results.mean())

print("\nİlk Ay Retention Oranları:")
print(cohort_results[1].sort_values(ascending=False))

# En Değerli 10 Müşteri
print("\nTop 10 Most Valuable Customers:")
print(ltv_results.nlargest(10, 'LTV')[['ContactId', 'LTV', 'CustomerLife', 'TotalRevenue']])

# Segment bazlı LTV analizi
segment_ltv = pd.merge(rfm_results[['ContactId', 'Customer_Segment']], 
                      ltv_results[['ContactId', 'LTV']], 
                      on='ContactId')

plt.figure(figsize=(10, 6))
sns.boxplot(data=segment_ltv, x='Customer_Segment', y='LTV')
plt.title('LTV Distribution by Customer Segment')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Aylık ortalama müşteri değeri trendi
df['Month'] = df['closedate'].dt.to_period('M')
monthly_customer_value = df.groupby('Month').agg({
    'contactid': 'nunique',
    'amount': 'sum'
}).reset_index()
monthly_customer_value['AvgCustomerValue'] = monthly_customer_value['amount'] / monthly_customer_value['contactid']

plt.figure(figsize=(12, 6))
plt.plot(monthly_customer_value['Month'].astype(str), 
         monthly_customer_value['AvgCustomerValue'], 
         marker='o')
plt.title('Average Customer Value Trend')
plt.xlabel('Month')
plt.ylabel('Average Value per Customer ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
def campaign_roi_analysis(df):
    # 1. Kampanya Bazlı Temel Metrikler
    campaign_metrics = df.groupby('campaign.name').agg({
        'amount': ['count', 'sum', 'mean'],  # işlem sayısı, toplam gelir, ortalama gelir
        'contactid': 'nunique'  # tekil müşteri sayısı
    }).round(2)
    
    campaign_metrics.columns = ['transaction_count', 'total_revenue', 'avg_transaction', 'unique_customers']
    campaign_metrics = campaign_metrics.reset_index()
    
    # Dönüşüm metriklerini hesapla
    campaign_metrics['revenue_per_customer'] = campaign_metrics['total_revenue'] / campaign_metrics['unique_customers']
    campaign_metrics['transactions_per_customer'] = campaign_metrics['transaction_count'] / campaign_metrics['unique_customers']
    
    # 2. Platform Bazlı Performans Analizi
    platform_metrics = df.groupby('platform__c').agg({
        'amount': ['count', 'sum', 'mean'],
        'contactid': 'nunique'
    }).round(2)
    
    platform_metrics.columns = ['transaction_count', 'total_revenue', 'avg_transaction', 'unique_customers']
    platform_metrics = platform_metrics.reset_index()
    
    # Görselleştirmeler
    
    # 1. Kampanya Gelir Karşılaştırması
    plt.figure(figsize=(12, 6))
    sns.barplot(data=campaign_metrics.sort_values('total_revenue', ascending=False).head(10), 
                x='campaign.name', y='total_revenue')
    plt.title('En Yüksek Gelir Getiren 10 Kampanya')
    plt.xticks(rotation=45)
    plt.xlabel('Kampanya')
    plt.ylabel('Toplam Gelir')
    plt.tight_layout()
    plt.show()

    # 2. Kampanya Bazlı Ortalama İşlem Değeri
    plt.figure(figsize=(12, 6))
    sns.barplot(data=campaign_metrics.sort_values('avg_transaction', ascending=False).head(10), 
                x='campaign.name', y='avg_transaction')
    plt.title('En Yüksek Ortalama İşlem Değerine Sahip 10 Kampanya')
    plt.xticks(rotation=45)
    plt.xlabel('Kampanya')
    plt.ylabel('Ortalama İşlem Değeri')
    plt.tight_layout()
    plt.show()

    # 3. Platform Performans Karşılaştırması
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    
    # İşlem Sayısı
    sns.barplot(data=platform_metrics, 
                x='platform__c', y='transaction_count', ax=axes[0,0])
    axes[0,0].set_title('Platform Bazlı İşlem Sayısı')
    axes[0,0].tick_params(axis='x', rotation=45)
    
    # Toplam Gelir
    sns.barplot(data=platform_metrics, 
                x='platform__c', y='total_revenue', ax=axes[0,1])
    axes[0,1].set_title('Platform Bazlı Toplam Gelir')
    axes[0,1].tick_params(axis='x', rotation=45)
    
    # Ortalama İşlem Değeri
    sns.barplot(data=platform_metrics, 
                x='platform__c', y='avg_transaction', ax=axes[1,0])
    axes[1,0].set_title('Platform Bazlı Ortalama İşlem Değeri')
    axes[1,0].tick_params(axis='x', rotation=45)
    
    # Tekil Müşteri Sayısı
    sns.barplot(data=platform_metrics, 
                x='platform__c', y='unique_customers', ax=axes[1,1])
    axes[1,1].set_title('Platform Bazlı Tekil Müşteri Sayısı')
    axes[1,1].tick_params(axis='x', rotation=45)
    
    plt.tight_layout()
    plt.show()

    # Özet İstatistikler
    print("\nKampanya Performans Özeti (İlk 10):")
    print(campaign_metrics.sort_values('total_revenue', ascending=False).head(10)
          [['campaign.name', 'total_revenue', 'avg_transaction', 'unique_customers', 'revenue_per_customer']])
    
    print("\nPlatform Performans Özeti:")
    print(platform_metrics)

    return campaign_metrics, platform_metrics

# Analizi çalıştır
campaign_metrics, platform_metrics = campaign_roi_analysis(df)
```
